# Pareto exploration (power-user)

Explore the architecture trade space with mixed-integer NSGA-II: satellite count, accelerators per satellite, altitude, and continuous drivers. Requires the `mdao` and `viz` extras.

```python
import orbitdc as odc
from orbitdc.optimize.pareto import pareto_nsga2_mixed

space = odc.load_scenario("../scenarios/orbital_1mw_inference.yaml")
front = pareto_nsga2_mixed(
    space,
    ["lcoc", "kg_per_kw"],
    ["n_satellites", "accelerators_per_satellite", "altitude_km", "downlink_gbps"],
    pop_size=24,
    n_gen=15,
)
print(f"{front.n_points} non-dominated designs")
```

```python
import plotly.graph_objects as go

fig = go.Figure(
    go.Scatter(
        x=front.f[:, 0],
        y=front.f[:, 1],
        mode="markers",
        text=[f"{int(p['n_satellites'])} sats" for p in front.points],
    )
)
fig.update_layout(title="Pareto front", xaxis_title="LCOC $/PFLOP-day", yaxis_title="kg/kW")
fig
```

```python
# The best LCOC design and its architecture.
best = min(range(front.n_points), key=lambda i: front.f[i, 0])
print("best LCOC design:", {k: round(v, 2) for k, v in front.points[best].items()})
```
