# Orbital vs terrestrial: a first comparison

Load a space scenario and an Earth baseline, run the delivered-compute waterfall, and read the binding constraints. The point is the diagnosis, not a single number.

```python
import orbitdc as odc

space = odc.load_scenario("../scenarios/orbital_1mw_inference.yaml")
earth = odc.load_scenario("../scenarios/earth_hyperscale_baseline.yaml")
result = odc.compare(space, earth)
print(result.summary())
```

```python
print(result.explain_binding_constraints())
```

```python
from orbitdc.viz import plot_cost_waterfall, plot_delivered_waterfall

plot_delivered_waterfall(result.space)
plot_cost_waterfall(result.space)
```

```python
from orbitdc.optimize import monte_carlo, tornado
from orbitdc.viz import plot_tornado

entries = tornado(space)
plot_tornado(entries, result.space.lcoc_per_pflop_day)

mc = monte_carlo(space, result.earth.lcoc_per_pflop_day, n=500, seed=0)
print(f"P(space beats Earth) = {mc.p_space_wins:.1%}")
```
