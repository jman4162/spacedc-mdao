# Quick start

Load a space scenario and an Earth baseline, compare them, and read the verdict and the binding constraints. The package is skeptical by default — it reports where a design fails.

```python
import orbitdc as odc

space = odc.load_scenario("../scenarios/orbital_1mw_inference.yaml")
earth = odc.load_scenario("../scenarios/earth_hyperscale_baseline.yaml")
result = odc.compare(space, earth)
print(result.summary())
```

```python
print(result.explain_binding_constraints())
```

Every default number is provenance-tagged, and the full evaluation is serializable.

```python
print("catalogs:", odc.list_catalogs())
ev = result.space
print("delivered fraction:", f"{ev.delivered_fraction:.1%}")
print("thermal bottleneck:", ev.thermal_bottleneck)
odc.export_report(result, "quick_start_report.md", fmt="md")
```
