# Monte Carlo uncertainty

Propagate the uncertain drivers (launch $/kg, solar W/kg, radiator mass, failure rate, utilization) and ask: under what assumptions does space beat Earth? Requires the `viz` extra for the fan chart.

```python
import orbitdc as odc
from orbitdc.viz import plotly_figures as pf

space = odc.load_scenario("../scenarios/orbital_1mw_inference.yaml")
earth = odc.load_scenario("../scenarios/earth_hyperscale_baseline.yaml")
result = odc.compare(space, earth)

mc = result.monte_carlo(n=1000, seed=0)
print(f"P(space beats Earth) = {mc.p_space_wins:.1%}")
print(f"LCOC p10/p50/p90 = {mc.lcoc_p10:,.0f} / {mc.lcoc_p50:,.0f} / {mc.lcoc_p90:,.0f}")
```

```python
pf.monte_carlo_fan(mc, result.earth.lcoc_per_pflop_day)
```

Which single assumptions flip the verdict? The thresholds report the value at which space matches Earth.

```python
from orbitdc.diagnostics import format_thresholds

for line in format_thresholds(result.thresholds()):
    print(line)
```
